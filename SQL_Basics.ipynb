Q9. What is the Difference Between a Primary Key and a Foreign Key?

ANS:
| **Primary Key**                                                            | **Foreign Key**                                                       |
| -------------------------------------------------------------------------- | --------------------------------------------------------------------- |
| Uniquely identifies each record in a table.                                | Creates a relationship between two tables.                            |
| Values must be **unique** and **cannot be NULL**.                          | Values **can be duplicated** and may be **NULL** (unless restricted). |
| Only **one primary key** is allowed per table (it can be a composite key). | A table can have **multiple foreign keys**.                           |
| Ensures **entity integrity**.                                              | Ensures **referential integrity**.                                    |
| Exists in the same table whose records it identifies.                      | References the **primary key** (or a unique key) of another table.    |


Q10. What Are Constraints in SQL and Why Are They Used?

ANS:
Constraints in SQL are rules applied to table columns to ensure the accuracy, validity, and integrity of the data stored in a database. They prevent invalid data from being inserted, updated, or deleted.

**Why Are Constraints Used?**
* To maintain data integrity.
* To prevent invalid or duplicate data.
* To enforce business rules.
* To maintain relationships between tables.
* To improve the reliability and consistency of the database.
